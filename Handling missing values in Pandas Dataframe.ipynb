```python
df2.isnull().sum().sum()
```

```python
df.isnull().sum().sum()
```

```python
df3= df.fillna(value = 5)
df3.head(5)
```

```python
### Filling  Null values with a previous values

df4 = df.fillna(method = "pad")
df4
```

```python
df4.isnull().sum().sum()          ### Now no null value is null
```

```python
df
```

```python
### Filling null values with the next values

df5 = df.fillna(method = "bfill")
df5
```

```python
df6 = df.fillna(method = "pad", axis=1)
df6
```

```python
df7 = df.fillna(method = "bfill", axis = 1)
df7
```

### Filling different values in Null in different columns

```python
df8 = df.fillna({'society': 'abcd',
              'balcony': 'defg'})
```

```python
df8
```

### Filling Null value with the mean of the column

```python
df9 = df.fillna(value=df['balcony'].mean())
df9
```

```python
df9 = df.fillna(value=df['balcony'].max())
df9
```

```python
df9 = df.fillna(value=df['balcony'].min())
df9
```
